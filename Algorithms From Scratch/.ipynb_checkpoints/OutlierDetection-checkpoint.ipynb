```python
import numpy as np

#mean
mean = 0
std_dev=10
num_samples=20
random_data = np.abs(np.random.normal(mean,std_dev,num_samples))
random_data = np.array([int(x) for x in random_data])
#print(random_data)

```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(15,10))
ax = plt.axes()
ax.scatter(np.arange(len(random_data)),random_data)

ax.set_title("Outlier Detection")
ax.set_ylabel("error count")
ax.set_xlabel("time")
```

```python
import seaborn as sns
sns.boxplot(x=random_data)

print ("mean",np.mean(random_data))
print ("median",np.median(random_data))
# Box and whisker diagrams
#mid line = median of data
# 50% of all data is inside blue box
# outliers = Q3 + 1.5*IQR
# IQR = inter quartile range
```
